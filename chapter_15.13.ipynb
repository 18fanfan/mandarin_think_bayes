# 章節 15：維度處理

## 15.13 討論

看不見的物種問題是一個活躍的研究領域，而且作者相信本章節的演算法是一個新穎的貢獻。所以在不到 200 頁的篇幅我們從基礎的機率到研究的研究前沿者。作者對此非常開心。

本書的目標是展示三個相關的概念：
- 貝葉斯思考：貝葉斯分析的基礎想法是用機率分佈來表示不確定的信念，用資料來更新這些機率分佈，並且用新的結果做預測或做決策。
  - 計算的方法：本書的前提是認為計算的方式比數學的方式容易了解貝葉斯分析，而且更容易使用可重複使用的構建塊來實作貝葉斯方法，這些構建塊可以重新編排以快速解決實際問題。
  - 迭代式建模：大部分實際問題包含建模選擇以及真實度與負責度之前的平衡。通常不可能提前知道哪些因素應該包含在模型中以及哪些因素可以被抽像出來。最好的方法是用迭代式的方式。從簡單的模型開始，再漸漸的增加複雜度，使用每個模型來驗證其他模型。

這些想法是容易轉換而且很有用；它們適用於科學和工程的各個領域的問題，從簡單的例子到當前的研究。如果你做到這點，您應該準備將這些工具應用於與您的工作相關的新問題。作者希望你覺得他們對你有幫助。
