## 章節 7：預測（Prediction）

### 7.7 討論

一如往常，本章的分析基於模型選擇，建模是一個迭代的過程。普遍來說，一開始先從簡單的方法產生一個近似的答案，再來找出可能的錯誤根源，找機會改進它。以本章的例子來說，我考慮了下列選項：

- 我的先驗分佈是建立在各隊每場的平均得分。但這個統計數據平均了所有的對手。針對特定的對手，我們可能期望更多的變化性。例如一個最佳攻擊的隊伍對上最差防守的隊伍，每場期望的得分數可能會超過平均值好幾個標準差。
- 我只用了前四場冠軍系列戰的資料。我也可以用兩隊常規比賽的數據。其中一個複雜的因素是賽季期間隊伍會因為球員交易或是受傷而發生變化。所以最好給予近期的比賽多些權重。
- 為了運用所有知道的資訊，我們可以用所有常規季賽的資料去預測各隊的得分率，可能透過估計一個對戰組合的附加因子，來加以調整。這個方式會更複雜，但依然可行。


對於第一個選項，我們可以用所有常規季賽的資料估計所有對戰組合的變化。感謝 Dirk Hoag 在 http://forechecker.blogspot.com/, 我可以拿到很多常規賽季的常規比賽得分結果（非驟死賽）。

隊伍在不同的區域，在常規季賽中只會對戰 1 到 2 次，所以我專注在對戰過 4-6 次的組合。每一對戰組合，我計算平均每場的得分，就是為了估計 λ ，接著畫出這些估計的分佈。

平均數再一次是 2.8 ，但標準差為 0.85，遠高於估算所有隊伍的估計標準差。

如果我們用較高的標準差在分析一次，棕熊隊贏得系列戰的機率是 80%，遠高於之前低標準差的 57%。所以事實證明結果對於先驗分佈的變化是很敏感的，這也是有道理的，畢竟我們處理的資料很少。

基於上述低標準差跟高標準差的結果差別，似乎很值得多花點時間來選擇正確的先驗分佈。
