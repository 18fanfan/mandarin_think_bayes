## 章節 5：發生比與分佈相加（Odds and Addends）

### 5.3 Oliver 之血

在 MacKay 的《 Information Theory, Inference, and Learning Algorithms 》書中提到：

> 有兩個人在犯罪現場留下血跡。一位嫌疑犯 Oliver 經過檢驗血型為 O 型。而遺留的兩種血跡分別是 O 型（當地常見的血型約 60%）與 AB 型（少見的血型 1%）。請問現場發現的血跡能夠支持 Oliver 為犯人的主張嗎？

要回答此問題，我們需要思考數據是支持（或反對）假設的證據的意義是什麼。直覺上，我們可以說在**看到資料後某假設的機率比沒看到資料之前更高**，則此資料是支持此假設的證據。

例如在餅乾問題中，先驗發生比是 1:1 ，或機率 50%。後驗發生比是 3:2，或是機率 60%。所以我們可以說香草餅乾是支持碗 1 的證據。

發生比形式的貝葉斯理論可以讓更精準的表達此直覺。再一次：

$$ o(H=A|D) = o(H=A)\frac{p(D|H=A)}{p(D|H=B)} $$

兩邊同除以 o(H=A) 則：

$$ \frac{o(H=A|D)}{o(H=A)} = \frac{p(D|H=A)}{p(D|H=B)} $$

等號左邊是後驗發生比與先驗發生比的比率。等號右邊是似然性的比率，又稱為貝葉斯因子（Bayes factor）。

如果貝葉斯因子：

1. 大於 1，意味資料較可能是來自假設 A 相較於假設 B， 之下。又因為貝葉斯因子大於 1，也意味著看到資料後假設 A 的勝率變大。
1. 小於 1，意味資料較不可能是來自假設 A 相較於假設 B，所以看到資料後的假設 A 的勝率變小。
1. 等於 1，則資料來自兩個假設的機會是一樣的，所以看到資料後的假設 A 的勝率不變。

回到 Oliver 之血問題。我們考慮兩種假設：

1. Oliver 是其中一個留下血跡的人
  - 因為 Oliver 為 O 型，所以機率為 1 ， AB 型任意選取的機率為 1%。所以 p(D=兩種血跡為 O 型和 AB 型|H=1) = 1*0.01
1. Oliver 不是留下血跡的人
  - O 型任意選取的機率為 60% ， AB 型任意選取的機率為 1%，並且兩個人可以互換。所以 p(D=兩種血跡為 O 型和 AB 型|H=2) = 2*0.6*0.01

所以貝葉斯因子為：

$$ \frac{p(D|H=1)}{p(D|H=2)} = \frac{1*0.01}{2*0.6*0.01} = \frac{1}{1.2} $$

貝葉斯因子小於 1 ，看起來資料是較為支持 Oliver 不是留下血跡之人的假設。

這是有點刻意的例子，但這是一個**違反直覺的例子**，即使資料與假設吻合不必然是支持假設的證據。

如果這違反直覺的結果很困擾你，這邊提供另一個想法：

資料是由一個**常見的血型 O**與**少見的血型 AB** 所構成。如果 Oliver 解釋了常見的事件，但是少見的事件任然沒有解釋。若 Oliver 不是被考慮的，則我們有兩次的機會在人群中找到一個 AB 型的人（因為兩個人可以交換）。這兩次的機會導致了這個差異。
